### 1. Generate three separate 1-dimensional arrays of 10,000 elements each.
* Array 1 should contain numbers randomly drawn from the uniform distribution between 1 and 100
* Array 2 should contain numbers randomly drawn from the normal distribution, with a mean of 0 and a standard deviation of 1
* Array 3 should contain numbers randomly drawn from the binomial distribution, where n=10 and p=0.5

```python
array1 = np.random.uniform(1, 100, 10000)
array1
```

```python
array2 = np.random.normal(0, 1, 10000)
array2
```

```python
array3 = np.random.binomial(10, 0.5, 10000)
array3
```

### 2. For each array, randomly sample 1,000 rows without replacement. (hint: `np.random.choice()`)


```python
array4 = np.random.choice(array1, size=1000)
array4
```

```python
array5 = np.random.choice(array2, size=1000)
array5
```

```python
arrary6 = np.random.choice(array3, size=1000)
arrary6
```

### 3. For each of the three samples:  
- Calculate the mean, median, and standard deviation
- Find the index and the value of the largest element (hint: `np.argmax()`)  
- Find the value of the 75th percentile (hint: `np.percentile()`)  

```python
# mean
mean1= array4.mean()
mean1

# median
median1 = np.median(array4)
median1

# STD
std1 = np.std(array4)
std1

# max
max1 = np.max(array4)
max1

max_index = np.argmax(array4)
max_index

# 3rd quartile
q3_1 = np.percentile(array4, 75)
q3_1
```

```python
# mean
mean2= array5.mean()
mean2

# median
median2 = np.median(array5)
median2

# STD
std2 = np.std(array5)
std2

# max
max2 = np.max(array5)
max2

max_index = np.argmax(array5)
max_index

# 3rd quartile
q3_2 = np.percentile(array5, 75)
q3_2
```

### 4. Sort each of the three samples in increasing order, then combine them together to create a 1,000 by 3 array. (hint: `np.sort()` and `np.stack()`). Now, check the shape of the array.

```python
array4_sorted = np.sort(array4)

array5_sorted = np.sort(array5)

array6_sorted = np.sort(arrary6)
```

```python
all_samples = np.stack((array4_sorted, array5_sorted, array6_sorted), axis=1)
all_samples
```

### 5. Replace all negative values with 0.

Replace all values that exceed 10 with 10.

```python
condition = all_samples < 0
condition
```

```python
all_samples[condition]
```

```python
all_samples_modified = np.clip(all_samples, 0, 10)
all_samples_modified
```

### 6. Subtract 5 from all values that exceed 5. (hint: `np.where()`)

```python
all_samples_modified = np.where(all_samples_modified > 5, all_samples_modified - 5, all_samples_modified)
all_samples_modified
```

### 7. Produce a fourth column that is the sum of the first two columns.  (hint: `np.hstack()`). Now check the new shape of the array.

```python
sum_first_two_columns = all_samples_modified[:, 0:2].sum(axis=1)
sum_first_two_columns.shape
sum_first_two_columns = sum_first_two_columns.reshape(1000, 1)
```

```python
final_sample = np.hstack((all_samples_modified, sum_first_two_columns))
final_sample
```

### 8. Use broadcasting to add 1 to each element of the first column, 2 to each element of the second column, 3 to each element of the third column, and 4 to each element of the fourth column.

```python
final_sample[:, 0] = final_sample[:, 0] + 1
final_sample[:, 1] = final_sample[:, 1] + 2
final_sample[:, 2] = final_sample[:, 2] + 3
final_sample[:, 3] = final_sample[:, 3] + 4

final_sample
```

### 9. Matrix multiply this 1000x4 array by a 4x1 array of ones. How many unique values are in the resulting product?  (hint: `np.unique()`)

```python
ones = np.ones((4, 1))

product = np.dot(final_sample, ones)
product

# product = final_sample @ ones

unique_values = np.unique(product)
unique_values.shape
```
